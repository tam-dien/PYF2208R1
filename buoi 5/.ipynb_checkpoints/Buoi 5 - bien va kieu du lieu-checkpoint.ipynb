# BIẾN

### Biến là gì???

Biến là một khái niệm trừu tượng nhưng vô cùng quan trọng trong lập trình

Biến là nơi để lưu trữ những giá trị tạm thời

Mỗi lần ta khai báo một biến, chương trình sẽ sử dụng một ít vùng nhớ trong RAM

Mỗi một biến sẽ mang một loại kiểu dữ liệu riêng

Có rất nhiều kiểu dữ liệu

### Khai báo biến trong Python

Khai báo biến trong python rất đơn giản theo cách sau:

`tên biến = giá trị biến`

Ví dụ:

`a = 2`

hoặc:

`thaydiendeptraithe = "HAHA"`

Kiểu dữ liệu của mỗi biến, sẽ tự được định dạng khi khai báo



##### Nguyên tắc đặt tên biến:
1. Không bắt đầu từ số
2. Biến trong python CÓ phân biệt chữ hoa và thường
3. Không dùng ký tự đặc biệt
4. Tên biến trong được trùng với những keyword có sẵn (if, else,...)

#### Một số kiểu dữ liệu trong python:
- Kiểu Number
- Kiểu String
- Kiểu Boolean
- Kiểu Tuple
- Kiểu List
- Kiểu Dictionary

### Kiểu số (Number)

Khai báo kiểu số:

`a = 5` (Kiểu số nguyên)

`a = 5.4` (Kiểu số thực)

Thực hiện một phép tính với kiểu số:

```python
a = 6
b = 7
print(a + b)
```

### Kiểu chuỗi (String)

Kiểu chuỗi là kiểu dùng để lưu trữ các ký tự (các chữ cái, con số, ký tư đặc biệt,...)

Khai báo kiểu chuỗi:

`a = "Hello mọi người"`

Thực hiện phép tính cộng với kiểu chuỗi:

```python
a = "6"
b = "7"
print(a + b)
```

### Kiểu boolean

Kiểu boolean là kiểu chỉ lưu trữ 2 trạng thái True (đúng) và False (sai)

Khai báo kiểu boolean:

`a = True`

### Chuyển đổi kiểu dữ liệu

Để chuyển đổi một kiểu dữ liệu này sang một kiểu dữ liệu khác, ta chỉ cần dùng lệnh:

`&kiểu dữ liệu&(&giá trị cần đổi&)`

Ví dụ:

`a = 2
b = str(a)`

Lúc này, biến a là kiểu số, b là kiểu chuỗi có giá trị là "2"

Hãy thử chuyển từ kiểu chuỗi sang kiểu số

#### Chuyển đổi không thành công:

```python
a = "A"
b = int(a)
```

#### Chuyển đổi thành công:

```python
a = "5"
b = int(a)
print(b)
```

# TOÁN TỬ

Trong lập trình một toán tử là một ký tự được sử dụng để thực hiện một phép toán số hoặc logic

Python hỗ trợ nhiều loại toán tử khác nhau

#### Các các tử phổ biến

- Toán tử số học
- Toán tử so sánh
- Toán tử gán giá trị (cho biến)
- Toán tử logic

### Toán tử số học

Python hỗ trợ tất cả các toán tử số học phổ biến dùng để thực hiện cộng, trừ, nhân, chia, luỹ thừa và tính phần dư



Thực hiện một phép chia lấy phần dư:

```python
a = 6
b = a % 4
print(b)
```

Thực hiện một phép tính bất kỳ rồi lưu vào biến:

```python
a = 67
b = 23
c = a + b
d = a - b
e = a / b
f = a % b
g = a * b
h = a // b
j = a**b
print(c)
print(d)
print(e)
print(f)
print(g)
print(h)
print(j)
```

### Toán tử so sánh

Khi thực hiện một toán tự so sánh, kết quả sẽ trả về là True (đúng) hoặc False (sai)



Thực hiện thử một phép so sánh:

```python
a = 5
b = 7
c = (a == b)
print(c)
```

### Toán tử gán giá trị

Toán tử gán giá trị dùng để thiết lập giá trị ban đầu hoặc thay đổi giá trị của biến



Lệnh `a += 3` là viết tắt của lệnh `a = a + 3`, đây được gọi là phép cộng dồn

Thực hiện thử một toán tử gián giá trị:

```python
a = 6
a += 3
print(a)
```

### Toán tử logic

Toán tự logic dùng để kết hợp các điều kiện

Toán tử logic trả về giá trị là True (đúng) hoặc False (sai)



Thực hiện thử một phép toán tử logic:

```python
a = 5
b = 6
c = 7
d = 8
e = (5 == 6) and (7 < 8)
f = (a < d) or (b > d)
h = ((a < d) or (b < d)) and (a < b)
g = (b < d) and not(b > d)
print(e)
print(f)
print(h)
print(g)
```

# Bài tập

Viết chương trình chuyển đổi nhiệt độ từ độ C của biến t sang độ F rồi in lên màn hình

```python
t = 35
f = ((t*9/5) + 32)
print(str(f) + " độ F")
```

Sử dụng thư viện Turtle để vẽ một tam giác đều với độ dài các cạnh là biến a

Sau đó tính chu vi và diện tích của hình tam giác đó

```python
import turtle

t = turtle.Turtle()

a = 50

t.fd(a)
t.left(120)
t.fd(a)
t.left(120)
t.fd(a)

chuvi = a*3

dientich = a*a*(3**(1/2))/4

print(chuvi)
print(dientich)
```

Sử dụng thư viện Turtle để vẽ một hình tròn với bán kính là biến r

Sau đó tính chu vi và diện tích của hình tròn đó

```python
import turtle
import math

t = turtle.Turtle()

r = 50

pi = 3.14

t.circle(r)

chuvi = 2*math.pi*r

dientich = r**2*math.pi

print(chuvi)

print(dientich)
```

Cho hệ phương trình sau:

a1x + b1y = c1

a2x + b2y = c2

Cho các biến a1, a2, b1, b2, c1, c2, tính x và y

```python
a1 = 3
b1 = 4
c1 = 5
a2 = 1
b2 = 2
c2 = 12
y = (a2*c1-c2)/(b1*a2/a1 -b2)
x = (c1 - b1*y)/a1
print(x)
print(y)
```

Cho biến a, hãy vẽ a hình vuông

```python
import turtle
cir = turtle.Turtle()
a = 7
for i in range(a):    
    for i in range(4):    
        cir.forward (30)
        cir.right(90)
    cir.penup()
    cir.forward(35)
    cir.pendown()
```
