### Lệnh if thiếu

Lệnh if là một trong ba lệnh rẽ nhánh cốt lõi của một ngôn ngữ lập trình

Cú pháp:

`if (điều kiện):
    //Câu lệnh 1
    //Câu lệnh 2
//Câu lệnh 3
`
    
Trong đó:
- Điều kiện là một biểu thức hoặc một biến có kiểu dữ liệu là boolean (kiểu chỉ chứa True hoặc False)

Các câu lệnh tab vào trong 1 tab so với lệnh if (câu lệnh 1, câu lệnh 2) là những lệnh nằm trong dòng if (chỉ thực hiện khi điều kiện đúng (True))

Các lệnh không tab vào trong (nằm cùng tab so với lệnh if - câu lệnh 3) là những lệnh KHÔNG nằm trong if, những lệnh này sẽ thực hiện sau khi những lệnh trong if thực hiện xong (nếu if đúng)

Sơ đồ tư duy của lệnh if:



Ví dụ:

Nhập vào một số a, kiểm tra xem a có phải là số lớn hơn 10 không, nếu có thì cho a thành 10, còn không thì giữ nguyên

In a

```python
a = float(input("Nhập vào số a:"))
if (a > 10):
    a = 10
print(a)
```

Nhập vào lương nhận tháng này, nếu lương trên 2 triệu thì trừ 10% thuế, dưới 2 triệu thì không trừ, sau đó in ra tiền lương thực lãnh:

```python
a = float(input("Nhập lương : "))
if a > 2000000:
    a = a*0.9
print(a)
```

### Lệnh if đủ (if... else)

Cú pháp:

`if (điều kiện):
    //Câu lệnh 1
    //Câu lệnh 2
else:
    //Câu lệnh 3
Câu lệnh 4`

Sơ đồ tư duy của lệnh if... else



Ví dụ:

Nhập vào một số, kiểm tra xem số vừa nhập vào có chia hết có 3 hay không, nếu có thì in ra "Đây là số chia hết cho 3", nếu không thì in ra "Đây không phải là số chia hết cho 3"

```python
a = int(input("Nhập vào một số a:"))
if (a % 3 == 0):
    print("Đây là số chia hết cho 3")
else:
    print("Đây không phải là số chia hết cho 3")
```

Nhập vào số điểm, nếu điểm nhập vào lớn hơn 0 và bé hơn 10 thì in ra dòng chữ "Đúng quy định", nếu không thì in ra dòng chữ "Sai quy định"

```python
a = float(input("Nhập vào điểm số:"))
if (a >= 0) and (a <= 10):
    print("Đúng quy định")
else:
    print("Sai quy định")
```

### Lệnh if lồng

Chỉ đơn giản if trong if

Cú pháp:

`if (điều kiện 1):
    //Câu lệnh 1
    if (điều kiện 2):
        //Câu lệnh 2
    //Câu lệnh 3
//Câu lệnh 4`

Sơ đồ tư duy của lệnh if lồng:



Ví dụ:

Nhập vào điểm toán và điểm văn, nếu cả 2 điểm đều nhập đúng theo quy định (nhập số từ 0 - 10 điểm), thì tiến hành tính và in ra điểm trung bình. Sau đó xét điểm trung bình, nếu điểm trung bình lớn hơn hoặc bằng 8, in ra dòng chữ "Học sinh giỏi", ngược lại, in ra dòng chữ "Không là học sinh giỏi".

Trong trường hợp nhập sai quy định, in ra dòng chữ "Nhập sai quy định"

```python
toan = float(input("Nhap vao diem toan: "))
van = float(input("Nhap vao diem van: "))
if (toan >= 0) and (toan <= 10) and (van >= 0) and (van <= 10):
    dtb = (toan + van)/2
    print(dtb)
    if (dtb >= 8):
        print("Học sinh giỏi")
    else:
        print("Không là học sinh giỏi")
else:
    print("Sai quy định")
```

### Lệnh if mở rộng (if...elif...else)

Mở rộng lệnh if tí

Cú pháp:

`if (điều kiện 1):
    //Câu lệnh 1
elif (điều kiện 2):
    //Câu lệnh 2
else:
    //Câu lệnh 3
//Câu lệnh 4`

Cú pháp trên tương đương với cú pháp:

`if (điều kiện 1):
    //Câu lệnh 1
else:
    if (điều kiện 2):
        //Câu lệnh 2
    else:
        //Câu lệnh 3
//Câu lệnh 4`

Sơ đồ tư duy của lệnh if mở rộng:



Ví dụ:

Nhập vào điểm số, nếu đúng quy định ta tiến hành xét:

Điểm >= 8: In ra học sinh giỏi

8 > điểm >= 6: In ra học sinh khá

6 > điểm >= 4: In ra học sinh trung bình

4 > điểm >= 2: In ra học sinh yếu

2 > điểm >= 0: In ra học sinh kém

```python
diem = float(input("Nhập vào điểm số:"))
if (diem >= 8 and diem <= 10):
    print("Học sinh giỏi")
elif (diem >= 6):
    print("Học sinh khá")
elif (diem >= 4):
    print("Học sinh trung bình")
elif (diem >= 2):
    print("Học sinh yếu")
elif (diem >= 0):
    print("Học sinh kém")
else:
    print("Nhập sai quy định")
```

Nhập vào một số, kiểm tra xem đó có phải là số chính phương hay không. Nếu có, hãy kiểm tra xem đó là bình phương của một số chẳn hay số lẻ.

```python
n = int(input("Nhập vào một số: "))
a = n**(1/2)
if n > 0 and (a % 1 == 0):
    if (a % 2 == 0 ):
        print("Số",n, "là số chính phương chẵn")
    else:
        print("Số",n, "là số chính phương lẻ")
else:
    print("Số",n, "không là số chính phương")
```

Lập trình chương trình đoán số như sau:

Người chơi nghĩ ra 1 số bất kỳ (từ 0 đến 100)

Máy tính sẽ tiến hành đoán số, mỗi lần máy tính đoán 1 số, người chơi hãy trả lời chính xác 1 trong 3 câu trả lời: lớn hơn, bé hơn hoặc bằng số người chơi nghĩ trong đầu

Sau tối đa 7 lần đoán, máy phải đoán đúng số mà người chơi đã nghĩ

```python
nguongtren = 100
nguongduoi = 0
sodoan = int((nguongtren - nguongduoi + 1)/2)
for i in range(7):
    print("Tui đoán số bạn chọn là",sodoan)
    phanhoi = input("Nếu sai, hãy nhập lớn hoặc bé tương ứng nếu số tui chọn lớn hoặc bé hơn số bạn nghĩ. Nếu đúng, hãy nhập đúng")
    if (phanhoi == "lớn"):
        nguongtren = sodoan - 1
        sodoan = nguongtren - int((nguongtren - nguongduoi + 2)/2) + 1
    elif (phanhoi == "bé"):
        nguongduoi = sodoan + 1
        sodoan = nguongduoi + int((nguongtren - nguongduoi + 2)/2) - 1
    elif (phanhoi == "đúng"):
        break
print("Vậy là tui đoán đúng số bạn nghĩ rồi nhé, số bạn nghĩ là:",sodoan)
```

Nhập vào 3 số a, b, c. Kiểm tra xem 3 số trên có thể tạo thành 1 tam giác hay không. Nếu không in ra "Không thể tạo thành tam giác", nếu có ta xét:

Nếu 3 cạnh trên tạo ra 1 tam giác thường, in ra "Đây là tam giác thường"

Nếu 3 cạnh trên tạo ra 1 tam giác đều, in ra "Đây là tam giác đều", vẽ tam giác đó ra

Nếu 3 cạnh trên tạo ra 1 tam giác vuông cân, in ra "Đây là tam vuông cân", vẽ tam giác đó ra

Nếu 3 cạnh trên tạo ra 1 tam giác vuông, in ra "Đây là tam vuông", vẽ tam giác đó ra

Nếu 3 cạnh trên tạo ra 1 tam giác cân, in ra "Đây là tam cân", vẽ tam giác đó ra

```python
import turtle
t = turtle.Turtle()
import math
a=float(input('nhập vào số a:'))
b=float(input('nhập vào số b:'))
c=float(input('nhập vào số c:'))


if a <= b <= c:
    maxx = c
    minn=a
    tb=b
elif a <= c <= b:
    maxx = b
    minn=a
    tb=c
    
elif b <= a <= c:
    maxx = c
    minn=b
    tb=a
elif b <= c <= a:
    maxx = a
    minn=b
    tb=c
elif c <= a <= b:
    maxx = b
    minn=c
    tb=a
else:
    maxx = a
    minn=c
    tb=b
     
 
        
        
print(maxx, minn, tb)

if maxx>(minn+tb) or a<=0 or b<=0 or c<=0:
    print('Đây không phải là tam giác')
    
else:
    if maxx==minn and tb==minn:
        print('Đây là tam giác đều')
        for i in range(3):
            t.forward(maxx)
            t.left(120)
    elif minn==tb and maxx==math.sqrt(minn**2+tb**2):
        t.forward(minn)
        t.left(90)
        t.forward(minn)
        t.left(135)
        t.forward(maxx)
        print('Đây là tam giác vuông cân')
    elif maxx==math.sqrt(minn**2+tb**2):
        t.forward(minn)
        t.left(90)
        t.forward(tb)
        t.goto(0,0)
        print('Đây là tam giác vuông')
    elif minn==tb:
        h=math.sqrt(minn**2-maxx/2**2)
        t.forward(maxx/2)
        t.goto(0,h)
        t.goto(-maxx/2,0)
        t.goto(0,0)
        print('Đây là tam giác cân')
    else:
        print('Đây là tam giác thường')
```
