# Thực hành List

##### Chú ý: list một dãy số người ta còn gọi là một mảng

### Bài 1

Viết hàm với đầu vào là một mảng, hãy tính và trả về giá trị trung bình của mảng

```python
def ham(L):
    S = 0
    for i in L:
        S += i
    return S/len(L)
    
L_so = [5,6,4,2,3,-1,-2]
print(ham(L_so))
```

### Bài 2

Viết hàm với đầu vào là một mảng, hãy tìm và trả về giá trị âm lớn nhất trong mảng

```python
def ham(L):
    L_am = []
    for i in L:
        if i < 0:
            L_am.append(i)
    m = L_am[0]
    for i in L_am:
        if m < i:
            m = i
    return m
L_so = [5,6,-3,4,-16,-2,-5,5]
print(ham(L_so))
```

```python
def L_TimSoAmLonNhat(List):
    Maxx = List[0]
    for i in List:
        if i < 0 and i > Maxx:
            Maxx = i
    return Maxx
List = [6,1,-2,8,-44,12,98]
print("Số âm lớn nhất trong List là:",L_TimSoAmLonNhat(List))
```

### Bài 3

Viết hàm với giá trị đầu vào là một mảng, hãy sắp xếp và trả về mảng có độ lớn tăng dần

```python
def sapxep(L):
    for a in range(len(L)-1):
        for i in range(a+1,len(L)):
            if L[a] > L[i]:
                L[a],L[i] = L[i],L[a]
    return L
    
L = [4,1,2,10,6,12,0]
L_sapxep = sapxep(L)
print(L_sapxep)
```

### Bài 4

Viết hàm với giá trị đầu vào là một mảng, tìm x sao cho đoạn [-x,x] chứa tất cả các giá trị có trong mảng

```python
def timx(L):
    m = abs(L[0])
    for i in L:
        if abs(i) > m:
            m = abs(i)
    return m

L = [4,1,2,-20,6,12,0]
print(timx(L))
```

### Bài 5

Viết hàm với giá trị đầu vào là một mảng, Hãy tìm một vị trí trong mảng thỏa hai điền kiện: có hai giá trị lân cận và giá trị tại vị trí đó bằng tích hai giá trị lân cận. Nếu mảng không tồn tại giá trị như vậy thì hàm trả về giá trị - 1

```python
def ham(L):
    for i in range(1,len(L)-1): ### nếu len(L) = 2 ~~> range(1,1)
        if L[i] == L[i-1]*L[i+1]:
            return L[i]
    return -1

L = [5,6,4,2,3,4,8,2]
print(ham(L))
```

### Bài 6

Viết hàm có giá trị đầu vào là một mảng. Hãy tìm giá trị đầu tiên trong mảng một chiều các số nguyên có chữ sốđầu tiên là chữ số lẻ. Nếu trong mảng không tồn tại giá trị như vậy thì hàm sẽ trả về giá trị 0.

```python
def ham(L):
    for i in L:
        i = str(i)
        if int(i[0])%2 == 1:
            return int(i)
    return 0

L = [4,2,12,13,71,75]
print(ham(L))
```

### Bài 7

Viết hàm có giá trị đầu vào là một mảng. Hãy tìm giá trị đầu tiên trong mảng có dạng 2^k. Nếu mảng không tồn tại giá trị dạng 2^k thì hàm sẽ trả về giá trị 0.

```python
def ham2muk(a):
    while a % 2 == 0:
        a = a//2
    if a == 1:
        return True
    return False

def ham(L):
    for i in L:
        if ham2muk(i):
            return i
    return 0

L = [5,4,2,1,32,16,4,17]
print(ham(L))
```

### Bài 8

Viết hàm có giá trị đầu vào là một mảng. Hãy tìm giá trị thỏa điều kiện toàn chữ số lẻ và là giá trị lớn nhất thỏa điều kiện ấy trong mảng (nếu mảng không có giá trị thỏa điều kiện trên thì hàm trả về giá trị 0)

```python
def toanchule(a):
    if a < 0:
        a *= -1
    for i in str(a):
        if int(i) % 2 == 0:
            return False
    return True
    
def ham(L):
    m = 0
    for i in L:
        if toanchule(i):
            if i > m or m == 0: ## khi m bằng 0 mà có số toàn lẻ thì cho m = 1
                m = i
    return m
L = [10000,-333,-777,-357,23,5234,2424]
print(ham(L))
```
